## Cleaning for the competitor data 


```python
import pandas as pd
import time
```

```python
filepath1 = "../Resources/All olymic competitors/athleteEventnocRegionsKEY.csv"
filepath2 = "../Resources/All olymic competitors/athleteEvents.csv"
filepath3 = "../Resources/All olymic competitors/Olympics_overall.csv"
df1 = pd.read_csv(filepath1)
df2 = pd.read_csv(filepath2)
df3 = pd.read_csv(filepath3)
```

```python
#get total athletes and medal count for each country and year
df2_g = df2.copy()
athlete_count = df2_g.groupby(["Year", "Team"])["ID"].nunique().reset_index(name="Total Athletes")
df2_g = df2_g.dropna(subset=["Medal"])
medal_count = df2_g.groupby(["Year", "Team"]).size().reset_index(name="Medal Count")
clean_df2 = pd.merge(athlete_count, medal_count, on=["Year", "Team"], how="left")
clean_df2["Medal Count"] = clean_df2["Medal Count"].fillna(0)
clean_df2.head(50)
```

```python
# aggregate age, height, weight by sex for each year and country
df2_s = df2.copy()
results_sex = []
start_time = time.time()

years = df2_s["Year"].unique()
teams = df2_s["Team"].unique()

for year in years:
    for team in teams:
        males = df2_s[(df2_s["Year"] == year) & (df2_s["Team"] == team) & (df2_s["Sex"] == "M")]
        females = df2_s[(df2_s["Year"] == year) & (df2_s["Team"] == team) & (df2_s["Sex"] == "F")]
        
        avg_age_m = males["Age"].mean()
        avg_height_m = males["Height"].mean()
        avg_weight_m = males["Weight"].mean()

        avg_age_f = females["Age"].mean()
        avg_height_f = females["Height"].mean()
        avg_weight_f = females["Weight"].mean()
        
        results_sex.append({
            "Year": year,
            "Team": team,
            "Average Age (M)": avg_age_m,
            "Average Height (M)": avg_height_m,
            "Average Weight (M)": avg_weight_m,
            "Average Age (F)": avg_age_f,
            "Average Height (F)": avg_height_f,
            "Average Weight (F)": avg_weight_f
        })

athletes_df = pd.DataFrame(results_sex)
total_time = time.time() - start_time
minutes = total_time/60
print(f" This took {total_time} seconds.... or {minutes} minutes....")
athletes_df.head(50)
```
